```python
from transformers import pipeline

model_path = "Qwen/Qwen3-4B-merged"

generator = pipeline(
    "text-generation", 
    model_path, 
    dtype="auto", 
    device_map="auto",
)
```

```python
messages = [
    {"role": "user", "content": "医生，我最近查出有血脂异常，这会不会导致动脉粥样硬化？"}
]
messages = generator(messages, max_new_tokens=256)[0]["generated_text"]
print(messages[-1]["content"])
```
